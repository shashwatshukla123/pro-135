```python
import matplotlib.pyplot as plt
import csv
```

```python
rows = []

with open("star_with_gravity.csv", "r") as f:
    reader = csv.reader(f)

    for row in reader:
        rows.append(row)

headers = rows[0]
star_data_rows = rows[1:]
```

```python
names = []
distances = []
masses = []
radiuses = []
gravity = []
```

```python
for star_data in star_data_rows:
    if len(star_data) == 6:
        names.append(star_data[1])
        distances.append(star_data[2])
        masses.append(star_data[3])
        radiuses.append(star_data[4])
        gravity.append(star_data[5])
```

```python
plt.bar(names, masses)
plt.title("Star Name vs Mass")
plt.xlabel("Name")
plt.ylabel("Mass")
plt.show()
```

```python
plt.bar(names, radiuses)
plt.title("Star Name vs Radius")
plt.xlabel("Name")
plt.ylabel("Radius")
plt.show()
```

```python
plt.bar(names, distances)
plt.title("Star Name vs Distance")
plt.xlabel("Name")
plt.ylabel("Distance")
plt.show()
```

```python
plt.bar(names, gravity)
plt.title("Star Name vs Gravity")
plt.xlabel("Name")
plt.ylabel("Gravity")
plt.show()
```
